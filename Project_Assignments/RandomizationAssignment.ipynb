```python
import random

def random_bottle_weight():

    return 9 if random.random() < 0.1 else 10
```

```python
probability_9_oz = 0.1
probability_10_oz = 0.9

expected_weight = (9 * probability_9_oz) + (10 * probability_10_oz)
print(f"The expected weight of a random pull is: {expected_weight} ounces")
```

```python
def test_expected_weight(n=1000):
    total_weight = 0
    for _ in range(n):
        total_weight += random_bottle_weight()
    return total_weight / n


avg_weight = test_expected_weight(1000)
print(f"Expected weight: 9.9")
print(f"Simulated average (1000 pulls): {avg_weight}")
print(f"Agreement: {abs(avg_weight - 9.9) < 0.1}")
```

```python
probability_of_finding_9_oz = 0.1
expected_pulls = 1 / probability_of_finding_9_oz

print(f"we expect to make {expected_pulls} pulls before finding the 9-ounce bottle.")
```

```python
def pulls_until_short():
    count = 0
    while True:
        count += 1
        if random_bottle_weight() == 9:
            return count
results = [pulls_until_short() for _ in range(1000)]
print(sum(results)/len(results))
```

```python
# Q3
import random

plates = ['pink']*6 + ['blue']*6 + ['black']*6 + ['green']*6

N = 100000
count_pink = 0
count_at_least_one = 0

for _ in range(N):

    if random.choice(plates) == 'pink':
        count_pink += 1


    draw4 = random.sample(plates, 4)
    if 'pink' in draw4:
        count_at_least_one += 1

print(" Simulated P(pink) =", count_pink/N)
print(" Simulated P(at least one pink among 4) =", count_at_least_one/N)
```

```python
# Q4
import random

plates = ['pink'] * 6 + ['blue'] * 6 + ['black'] * 6 + ['green'] * 6
random.shuffle(plates)

plate_stack = plates.copy()

print("Q4:\n LIFO plate stack 'plate_stack' created and initialized.")
print(f"Current stack size: {len(plate_stack)}")
print(f"Top plate (last in): {plate_stack[-1] if plate_stack else 'Empty'}")
```

```python
#Q5.
import random

initial_plates = ['pink']*6 + ['blue']*6 + ['black']*6 + ['green']*6
random.shuffle(initial_plates)
plate_stack = initial_plates.copy()

dirty_plates = []

dishwasher_capacity = 12

def wash_and_return_plates(stack, dirty, capacity):
    random.shuffle(dirty)
    stack.extend(dirty)
    dirty.clear()

def draw_plates_for_dinner(stack, dirty, capacity, num_to_draw=4):
    if len(stack) < num_to_draw:
        wash_and_return_plates(stack, dirty, capacity)

    if len(stack) < num_to_draw:
        actual_num_to_draw = len(stack)
    else:
        actual_num_to_draw = num_to_draw

    drawn_plates = []
    for _ in range(actual_num_to_draw):
        if stack:
            drawn_plates.append(stack.pop())
        else:
            break
    return drawn_plates

num_days = 10000

exactly_one_pink_count = 0
at_least_one_pink_count = 0
total_draws = 0

for day in range(num_days):
    drawn_plates = draw_plates_for_dinner(plate_stack, dirty_plates, dishwasher_capacity)

    if len(drawn_plates) == 4:
        total_draws += 1
        pink_count = drawn_plates.count('pink')

        if pink_count == 1:
            exactly_one_pink_count += 1

        if pink_count >= 1:
            at_least_one_pink_count += 1

    dirty_plates.extend(drawn_plates)

    if len(dirty_plates) >= dishwasher_capacity:
        wash_and_return_plates(plate_stack, dirty_plates, dishwasher_capacity)

prob_exactly_one_pink = exactly_one_pink_count / total_draws if total_draws > 0 else 0
prob_at_least_one_pink = at_least_one_pink_count / total_draws if total_draws > 0 else 0

print(f" Q5:\nProbability of drawing exactly one pink plate: {prob_exactly_one_pink:.4f}")
print(f" Probability of drawing at least one pink plate: {prob_at_least_one_pink:.4f}")

print(" Process completed.")
```

```python
# Q6
def rejection_pull(stack):
    rejected = []
    while True:
        plate = stack.pop()
        if plate == 'pink':
            # return rejected plates
            stack.extend(rejected)
            return plate
        else:
            rejected.append(plate)

print("Q6:  Rejection pull result:", rejection_pull(plates.copy()))
```

```python
# Q7
stack = plates.copy()
random.shuffle(stack)
all_pink_days = 0

for day in range(1000):

    if day % 12 == 0:
        random.shuffle(stack)


    dinner = []
    while 'pink' not in dinner and stack:
        dinner = [stack.pop() for _ in range(min(4, len(stack)))]


    if all(p=='pink' for p in dinner):
        all_pink_days += 1


    stack.extend(dinner)

print(f"Q7:\nDays with all pink : {all_pink_days}/1000 = {all_pink_days/10:.1f}%")
```

```python
# Q8
from collections import Counter

def plate_usage():
    stack = plates.copy()
    random.shuffle(stack)
    use_count = Counter()

    for _ in range(1000):
        pulled = []

        for _ in range(4):
            p = rejection_pull(stack)
            pulled.append(p)
            use_count[p] += 1


        random.shuffle(pulled)
        stack.extend(pulled)

    return use_count

usage_results = plate_usage()
print("Q8:\n Usage counts:", usage_results)
```
